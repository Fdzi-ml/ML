```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import  pandas as pd
```

```python
data=pd.read_csv("Admission_Predict_Ver1.1.csv")
```

```python
data
```

```python
data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler as mm
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
X=data.drop(columns=['Chance of Admit '])
```

```python
X=mm().fit_transform(X)
```

```python
data.shape
```

```python
y=data[['Chance of Admit ']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=14,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_,lr.intercept_
```

```python
y_pred=lr.predict(x_test)
```

```python
plt.scatter(np.linspace(0.3,1,100),np.linspace(0.3,1,100),c='red')
plt.scatter(y_test,y_pred,alpha=0.5)
plt.xlabel("Actual Admit chance")
plt.ylabel('predicted value')
```

```python
from sklearn.linear_model import Ridge as ri
model=ri(alpha=0.3)
model.fit(x_train,y_train)
model.coef_,model.intercept_
```

```python
y_pred_r=model.predict(x_test)
plt.scatter(np.linspace(0.3,1,100),np.linspace(0.3,1,100),c='red')
plt.scatter(y_test,y_pred_r,alpha=0.5)
plt.xlabel("Actual Admit chance")
plt.ylabel('predicted value')
```

```python
from sklearn.linear_model import Lasso 
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.coef_,ls.intercept_
```

```python
y_pred_ls=ls.predict(x_test)
plt.scatter(np.linspace(0.3,1,100),np.linspace(0.3,1,100),c='red')
plt.scatter(y_test,y_pred_r,alpha=0.5)
plt.xlabel("Actual Admit chance")
plt.ylabel('predicted value')
```

```python

from sklearn.linear_model import ElasticNet as en
elastic=en(alpha=0.01,l1_ratio=0.05)
elastic.fit(x_train,y_train)
elastic.coef_,elastic.intercept_
```

```python
y_pred_el=elastic.predict(x_test)
plt.scatter(np.linspace(0.3,1,100),np.linspace(0.3,1,100),c='red')
plt.scatter(y_test,y_pred_r,alpha=0.5)
plt.xlabel("Actual Admit chance")
plt.ylabel('predicted value')
```

```python
from sklearn.preprocessing import PolynomialFeatures as pf
from sklearn.pipeline import Pipeline as pl
reg1=pl(steps=[("polynominal",pf(degree=2,interaction_only=False)),("regression",lr)])
reg1.fit(x_train,y_train)
```

```python
y_pred_re1=reg1.predict(x_test)
plt.scatter(np.linspace(0.3,1,100),np.linspace(0.3,1,100),c='red')
plt.scatter(y_test,y_pred_r,alpha=0.5)
plt.xlabel("Actual Admit chance")
plt.ylabel('predicted value')
```

```python
from sklearn.preprocessing import PolynomialFeatures as pf
from sklearn.pipeline import Pipeline as pl
reg1=pl(steps=[("polynominal",pf(degree=3,interaction_only=True)),("regression",lr)])
reg1.fit(x_train,y_train)
y_pred_re=reg1.predict(x_test)
plt.scatter(np.linspace(0.3,1,100),np.linspace(0.3,1,100),c='red')
plt.scatter(y_test,y_pred_r,alpha=0.5)
plt.xlabel("Actual Admit chance")
plt.ylabel('predicted value')
```

```python
from sklearn.metrics import max_error,r2_score,mean_absolute_error,mean_squared_error
```

```python
for j in [y_pred,y_pred_ls,y_pred_r,y_pred_re,y_pred_re,y_pred_el]:
    print("model",end='\n')
    for i in [max_error,r2_score,mean_absolute_error,mean_squared_error]:
        print(i(y_true=y_test,y_pred=j),end='\n')
```
